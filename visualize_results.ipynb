## Visualization of RoPE Parallelization Results

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
with open("serial_comparison.csv", "r") as f:
    data = pd.read_csv(f)


parallel_trials = data[ data['kernel'] == "parallel_128"]
sequential_trials = data[ data['kernel'] == "sequential"]
```

## Parallel vs Sequential (increasing matrix sizes)

### Runtime

```python
# Runtime Comparison
plt.plot(parallel_trials['seq_len'], parallel_trials['elapsed_ms'], label="parallel")
plt.plot(sequential_trials['seq_len'], sequential_trials['elapsed_ms'], label="sequential")
plt.xlabel("Sequence Length")
plt.ylabel("Elapsed Time (ms)")
plt.yscale("log")
plt.legend()
plt.show()
```

### Memory Bandwidth

```python
# Memory Comparison
plt.plot(parallel_trials['seq_len'], parallel_trials['bandwidth_gb_s'], label="parallel")
plt.plot(sequential_trials['seq_len'], sequential_trials['bandwidth_gb_s'], label="serial")
plt.xlabel("Sequence Length")
plt.ylabel("Memory Bandwidth (GB/s)")
plt.yscale("log")
plt.legend()
plt.show()
```

### Compute FLOPS

```python
# Compute Comparison
plt.plot(parallel_trials['seq_len'], parallel_trials['gflops'], label="parallel")
plt.plot(sequential_trials['seq_len'], sequential_trials['gflops'], label="serial")
plt.xlabel("Sequence Length")
plt.ylabel("Compute Performance (GFLOPS)")
plt.yscale("log")
plt.legend()
plt.show()
```
